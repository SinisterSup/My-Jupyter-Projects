```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open('DataSets/amazon_jobs_dataset.csv',encoding = "utf-8") as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    file_list = list(file_data)

count = [0 for i in range(12)]

for row in file_list:
    month = row['Posting_date'].split('-')[1]
    
    if 'Jan' in month:
        count[0] = count[0] + 1
    elif 'Feb' in month:
        count[1] = count[1] + 1
    elif 'Mar' in month:
        count[2] = count[2] + 1
    elif 'Apr' in month:
        count[3] = count[3] + 1
    elif 'May' in month:
        count[4] = count[4] + 1
    elif 'Jun' in month:
        count[5] = count[5] + 1
    elif 'Jul' in month:
        count[6] = count[6] + 1
    elif 'Aug' in month:
        count[7] = count[7] + 1
    elif 'Sep' in month:
        count[8] = count[8] + 1
    elif 'Oct' in month:
        count[9] = count[9] + 1
    elif 'Nov' in month:
        count[10] = count[10] + 1
    elif 'Dec' in month:
        count[11] = count[11] + 1
    
mnth = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July','August','September','October','November','December']

plt.plot(mnth, count)
plt.xlabel("Month")
plt.ylabel("No. of Job Postings")
plt.xticks(rotation=30)
plt.grid()
plt.show()

for i in range(len(count)):
    print(mnth[i],count[i])
```
